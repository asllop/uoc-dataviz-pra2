```python
def my_aggregator(a):
    if (len(a) > 0):
        return a.mode()
    else:
        return None
mode_all_years = final_df.groupby(['iso_a3'],as_index=False).agg(my_aggregator)
mode_all_years
```

```python
mean_all_years
```

```python
mean_all_years = mean_all_years.merge(mode_all_years, how='inner', on='iso_a3', copy=True)
```

```python
mean_all_years['IncomeGroup'] = mean_all_years['IncomeGroup'].apply(lambda x: str(x)).apply(lambda x: np.nan if (x == '[]') else x)
mean_all_years['IncomeGroup'].unique()
```

```python
print("Mitjana de tots els anys (1990-2017)")
colors = {'Low income':'red', 'Lower middle income':'orange', 'Upper middle income':'green','High income':'blue',np.nan:'black'}
print('Legend: ' + str(colors))
mean_all_years.plot.scatter(x='GDP', y='Air pollution (total) (deaths per 100,000)', c=mean_all_years['IncomeGroup'].map(colors))
```

```python
print("Grup d'ingressos per a més de 50 morts")
mean_all_years[mean_all_years['Air pollution (total) (deaths per 100,000)'] > 50].groupby(['IncomeGroup'],as_index=False).mean().plot.bar(x='IncomeGroup',y='Air pollution (total) (deaths per 100,000)')
```

```python
print("Grup d'ingressos per a menys de 50 morts")
mean_all_years[mean_all_years['Air pollution (total) (deaths per 100,000)'] <= 50].groupby(['IncomeGroup'],as_index=False).mean().plot.bar(x='IncomeGroup',y='Air pollution (total) (deaths per 100,000)')
```

# Generem dades de sortida

```python
sortida = {}
country_codes = final_df['iso_a3'].unique()
# Convert al NaN
final_df['GDP'] = final_df['GDP'].fillna(0)
final_df['IncomeGroup'] = final_df['IncomeGroup'].fillna('Unknown')
# Generem keys amb els anys
for year in range(1990, 2018):
    countries_per_year = {}
    for country in country_codes:
        obj = final_df[(final_df['iso_a3'] == country) & (final_df['Year'] == year)].to_dict(orient='records')[0]
        del obj['iso_a3']
        del obj['Entity']
        del obj['Year']
        countries_per_year[country] = obj
    sortida[str(year)] = countries_per_year
    
# Generem els països
sortida
```

```python
sortida['2001']['FRA']
```

```python
import json
f = open("dataset.json", "w")
json.dump(sortida, f)
f.close()
```

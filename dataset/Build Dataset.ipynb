```python
# Filtrem els valors nuls de codi de país
pollution_df = pollution_df[~pollution_df["Code"].isnull()]
pollution_df
```

```python
# Filtrem el codi "OWID_WRL" que no és cap país
pollution_df = pollution_df[pollution_df["Code"] != "OWID_WRL"]
pollution_df
```

```python
country_code = pollution_df["Code"].unique()
country_code
```

```python
# Filtra "income" pels països que tenim a "pollution"
income_df = income_df[income_df["Country Code"].isin(country_code)][["Country Code", "Region", "IncomeGroup"]]
income_df
```

```python
# Filtra GDP pels països que tenim a "pollution"
gdp_df = gdp_df[gdp_df["Country Code"].isin(country_code)]
gdp_df
```

```python
max_year = pollution_df["Year"].max()
max_year
```

```python
years = list(range(min_year, max_year + 1))
years
```

```python
# Fem una copia del dataaset de base per a no trencar res
final_df = pollution_df.copy()
```

```python
# Afegim columna Region al dataset final
tmp_regions = []
for code in final_df['Code']:
    region = income_df[income_df["Country Code"] == code]["Region"].to_list()
    if len(region) > 0:
        tmp_regions.append(region[0])
    else:
        # La regió de TWN és buida, l'omplim a ma
        if code == "TWN":
            tmp_regions.append("East Asia & Pacific")
        else:
            tmp_regions.append("")

final_df["Region"] = tmp_regions
final_df
```

```python
# Afegim columna IncomeGroup al dataset final
tmp_income_grp = []
for code in final_df['Code']:
    income_grp = income_df[income_df["Country Code"] == code]["IncomeGroup"].to_list()
    if len(income_grp) > 0:
        tmp_income_grp.append(income_grp[0])
    else:
        # El grup de TWN és buit, l'omplim a ma
        if code == "TWN":
            tmp_income_grp.append("High income")
        else:
            tmp_income_grp.append("")

final_df["IncomeGroup"] = tmp_income_grp
final_df
```

```python
# Afegim columna GDP

gdp_list = []
for index, row in final_df.iterrows():
    country = row["Code"]
    year = str(row["Year"])
    gdp = list(gdp_df[gdp_df["Country Code"] == country][year])
    if len(gdp) > 0:
        gdp_list.append(gdp[0])
    else:
        gdp_list.append(0)

final_df["GDP"] = gdp_list

final_df
```

```python
# Comprovem que tenim les dades correctes després de generar el dataset. Per exemple, Espanya l'any 2015

val_from_final = final_df[(final_df["Code"] == "ESP") & (final_df["Year"] == 2015)]["GDP"].values[0]
val_from_gdp = gdp_df[gdp_df["Country Code"] == "ESP"][["2015"]].values[0][0]

val_from_final == val_from_gdp
```

```python
# Generem nou CSV

final_df.to_csv("FINAL.csv", index=False)
```
